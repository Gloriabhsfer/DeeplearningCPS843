```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
sess = tf.InteractiveSession()
```

```python
import numpy as np
import math 
```

```python
X =np.expand_dims(np.arange(0.0, 5.0, 0.01),1)
Y =np.sinc(5*X)
x = tf.placeholder(tf.float64, [500,1], name='x')
y = tf.placeholder(tf.float64, [500,1], name='y')
```

insert the data type for your tensor

```python
input_layer = tf.layers.dense(x, 200, activation= tf.nn.relu)
hidden_layer1 = tf.layers.dropout(input_layer,0.2)
hidden_layer2 = tf.layers.dense(hidden_layer1,250,activation=tf.nn.tanh)
output_layer = tf.layers.dense(hidden_layer2,1)
Loss =tf.losses.mean_squared_error(y , output_layer)
Optimizer = tf.train.AdamOptimizer(learning_rate= 0.001).minimize(Loss)
init = tf.global_variables_initializer()
```

```python
loss_list=[]
sess.run(init)
for i in range(0,200):
  fd ={x:X, y:Y}
  _, loss_val = sess.run([Optimizer, Loss], feed_dict=fd)
  #print ('loss = %s' % loss_val)
  loss_list.append(loss_val)

YP = sess.run(output_layer,feed_dict={x:X})
```

```python

# Plot training  loss values
import matplotlib.pyplot as plt
plt.plot(loss_list)
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()
```

```python
plt.plot(X,Y)
plt.plot(X,YP)
plt.show()
```

```python
input_layer_m2 = tf.layers.dense(x, 200, activation= tf.nn.relu)
hidden_layer1_m2 = tf.layers.dropout(input_layer_m2,0.2)
hidden_layer2_m2 = tf.layers.dense(hidden_layer1_m2,250,activation=tf.nn.relu)
hidden_layer3_m2 = tf.layers.dense(hidden_layer2_m2,250,activation=tf.nn.relu)
hidden_layer4_m2 = tf.layers.dense(hidden_layer3_m2,500,activation=tf.nn.relu)
output_layer_m2 = tf.layers.dense(hidden_layer4_m2,1)
Loss_m2 =tf.losses.mean_squared_error(y , output_layer_m2)
Optimizer_m2 = tf.train.AdamOptimizer(learning_rate= 0.001).minimize(Loss_m2)
init_m2 = tf.global_variables_initializer()
```

```python
loss_list_m2=[]
sess.run(init_m2)
for i in range(0,200):
  fd ={x:X, y:Y}
  _, loss_val_m2 = sess.run([Optimizer_m2, Loss_m2], feed_dict=fd)
  #print ('loss = %s' % loss_val)
  loss_list_m2.append(loss_val_m2)

YP_m2 = sess.run(output_layer_m2,feed_dict={x:X})
```

```python
# Plot training  loss values
import matplotlib.pyplot as plt
plt.plot(loss_list)
plt.plot(loss_list_m2)
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()
```

```python
plt.plot(X,Y)
plt.plot(X,YP)
plt.plot(X,YP_m2)
plt.show()
```
